# Detector de bordas de Canny

Nesse tutorial veremos como utiliar o detector de bordas de Canny para extrair as arestas de uma imagem. Começamos abrindo a nossa imagem em escala de cinza. A imagem utilizada pode ser encontrada [aqui](https://drive.google.com/file/d/1Y-5Ji6GuswcpfMID58hKEUdiWW7mO-4A/view?usp=sharing).
```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

image = cv.imread("/content/drive/MyDrive/ColabFiles/messi.jpg")

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

%matplotlib inline

plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
```

O detector de Canny é implementado pelo método `Canny`. Esse método possui os seguintes parâmetros:
* A imagem de entrada
* O limiar inferior (histerese)
* O limiar superior (histerese)
* O tamanho do *kernel* para o gradiente (Sobel) (opcional, por *default* é 3)
* Um valor *booleano* indicando se a magnitude do gradiente será calculada pela norma L2 (opcional, por *default* é `false` e a magnitude é calculada pela norma L1)

O retorno do método é a imagem com as bordas.

```python
borders = cv.Canny(gray,100,200,apertureSize = 3,L2gradient = False)

plt.imshow(borders, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
```
